## Q.1. Calculate Contractual Margin based on Contractual Margin Type, TV and Contractual Margin Information. Preferably use Python to complete this task. 

## **Calculate Contractual Margin For Contractual Margin Type 1**

**Adding a new column named CM_Type_1 to the dataframe.**<br>
Fault Percentage of TV. Let's say  Eg : TV = 10,000, Contractual Margin Type = 1, Contractual Margin Information = 8. 

`Contractual Margin = 10,000 * 0.08 = 800`
    

```python
# Data['Contractual Margin Information'] = Data['Contractual Margin Information'].astype("int32")
```

```python
def cm_type_1(data):
    
    if data['Contractual Margin Type'] == 1:
        
        return (data['TV']) * (int(data['Contractual Margin Information'])/100)
    else:
        return 0
```

```python
Data['CM_Type_1'] = Data.apply(cm_type_1, axis = 1)
Data
```

## **Calculate Contractual Margin For Contractual Margin Type 2**

**Type 2** <br>
contractual margin is a flat fee. This number is not dependent on TV. <br>
Eg : TV = 10,000, Contractual Margin Type = 2, Contractual Margin Information = 2000.<br>

Contractual Margin = 2000


```python
def cm_type_2(data):
    
    if data['Contractual Margin Type'] == 2:
        
        return (int(data['Contractual Margin Information']))
    else:
        return 0
```

```python
Data['CM_Type_2'] = Data.apply(cm_type_2, axis = 1)
Data
```

## **Calculate Contractual Margin For Contractual Margin Type 3**

**Type 3**

contractual margin is a tier based percentage of TV. <br>

Eg :

Contractual Margin Type = 3, 

Contractual Margin Information = 0-75000(5),75000-150000(7),150000-999999(10).

If TV is greater than 0 and less than or equal to 75000 then 5% of TV,<br>
if TV is greater than 75000 and less than or equal to 150000 then 7% of TV,<br> 
if TV is greater than 150000 then 10% of TV.<br>

**If TV = 50,000 then Contractual Margin = 50,000 * 0.05 = 2500**


```python
# np.min(np.arange(0,3001)) <= row['Spend'] <= np.max(np.arange(0,3001)):
```

```python
def cm_type_3(data):
    
    if data['Contractual Margin Type'] == 3:
        
        if np.min(np.arange(0,75000))<=data['TV'] <= np.max(np.arange(0,75000)):
        
            return data['TV'] * 0.05
    
        elif np.min(np.arange(75000,150000))<=data['TV'] <= np.max(np.arange(75000,150000)):
        
            return data['TV'] * 0.07
        
        elif data['TV']> 150000:
                       
            return data['TV'] * 0.10
        
            
    else:
        return 0
```

```python
Data['CM_Type_3'] = Data.apply(cm_type_3, axis = 1)
Data
```

## **Calculate Contractual Margin For Contractual Margin Type 4**

**Type 4** is an incremental percentage fee.

Eg :

Contractual Margin Type = 4, 

Contractual Margin Information = 0-20000(7.5),20000-70000(6),70000-999999(4).<br>

If TVis greater than 0 and less than or equal to 20000 then contractual margin is 7.5% of TV.<br>
If TV is greater than 20000 and less than or equal to 70000 then contractual margin is 7.5% of 2000 + 6% of (TV-20000).<br>
If TV is greater than 70000 then contractual margin 7.5% of 2000 + 6% of 50000 + 4% of TV - 70000.<br>
If TV = 15,000 then Contractual Margin = 15,000 * 0.075 = 1125<br>
If TV = 40,000 then Contractual Margin = 20,000 * 0.075 + (40,000-20,000) *0.06 = 2700.<br> 
If TV = 80,000 then Contractual Margin = 20,000 * 0.075 + (70,000-20,000) * 0.06 + (80,000-70,000)*0.04 = 4900. 


```python
def cm_type_4(data):
    
    if data['Contractual Margin Type'] == 4:
        
        if np.min(np.arange(0,20000))<=data['TV'] <= np.max(np.arange(0,20000)):
        
            return data['TV'] * 0.075
    
        elif np.min(np.arange(20000,70000))<=data['TV'] <= np.max(np.arange(20000,70000)):
        
            return (20000 * 0.75) + (data['TV']-20000)*0.06
        
        else:
        
            return (20000 * 0.75) + (70000-20000) * 0.06 + (data['TV']-70000) *0.04
    else:
        return 0
```

```python
Data['CM_Type_4'] = Data.apply(cm_type_4, axis = 1)
Data
```

## **Calculate Contractual Margin For Contractual Margin Type 5**

**Type 5 is a combination of flat and percentage fee**<br>

Eg : 
    Contractual Margin Type = 5,<br>
    Contractual Margin Information = 0-5000(5),1500.<br>
    If TV is greater than 0 and less than 5000 then 5% and if TV is greater than 5000 then it is a flat fee of 1500.<br>
    If TV = 4000 then Contractual Margin = 4000 * 0.05 = 200.<br>
    If TV = 7500 then Contractual Margin = 2000.

```python
def cm_type_5(data):
    
    if data['Contractual Margin Type'] == 5:
          
        if np.min(np.arange(0,5000))<=data['TV'] <= np.max(np.arange(0,5000)):
        
            return data['TV'] * 0.05
    
        elif 5000<=data['TV']:
        
            return 1500
    else:
        return 0
```

```python
Data['CM_Type_5'] = Data.apply(cm_type_5, axis = 1)
Data
```

## Q.2.Calculate GR Forecast Delta % using the following formula :GR Forecast Delta % = (GR - GR Forecast) * 100 / GR Forecast

## **GR Forecast Delta %**

```python
def GR_forest(data):
    
    GR_forest_percentage = round((data['GR']-data['GR Forecast']) *100/(data['GR Forecast']),2)
    return GR_forest_percentage
```

```python
Data['GR Forecast Delta'] = Data.apply(GR_forest, axis = 1)
Data
```
